```python
## first instruction A, use keras library to build a neural network
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
#from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np
import pandas as pd
```

## A. Build Neural Network as a function

```python
concrete_data = pd.read_csv('concrete_data.csv')  ## running locally
## concrete_data = pd.read_csv('https://cocl.us/concrete_data') --- pulling data from internet
```

```python
#concrete_data.describe()
```

```python
#concrete_data.head()
predictors = concrete_data[concrete_data.columns[concrete_data.columns != 'Strength']]
target = concrete_data['Strength']
```

## B. Normalize the data (5 marks) 
Repeat Part A but use a normalized version of the data. Recall that one way to normalize the data is by subtracting the mean from the individual predictors and dividing by the standard deviation.
How does the mean of the mean squared errors compare to that from Step A?

```python
predictors_norm = (predictors - predictors.mean()) / predictors.std()
n_cols = predictors_norm.shape[1] # number of predictors
```

```python
def simple_network():
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
                    
    model.compile(optimizer='adam',
            loss='mean_squared_error',
            metrics=['accuracy'])
    return model

model = simple_network()
```

```python
MSE = []
```

```python
for v in range(51):
    x_train, x_test, y_train, y_test = train_test_split(predictors_norm, target, test_size=0.3)
    model.fit(x_train, y_train, epochs=50, verbose=2)
    y_pred = model.predict(x_test)
    mse = mean_squared_error(y_test, y_pred)
    MSE.append(mse)
```

```python
print('Mean: {}', 'Std_DevL {}, of MSE~ ', np.mean(MSE),',' ,np.std(MSE))
```
