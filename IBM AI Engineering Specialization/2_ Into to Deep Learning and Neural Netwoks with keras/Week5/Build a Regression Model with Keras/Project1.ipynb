```python
## first instruction A, use keras library to build a neural network
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
#from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np
import pandas as pd
```

```python
concrete_data = pd.read_csv('concrete_data.csv')  ## running locally
## concrete_data = pd.read_csv('https://cocl.us/concrete_data') --- pulling data from internet
```

```python
#concrete_data.describe()
```

```python
#concrete_data.head()
predictors = concrete_data[concrete_data.columns[concrete_data.columns != 'Strength']]
target = concrete_data['Strength']
```

```python
#predictors[:5]
#target[:5]
n_cols = predictors.shape[1] # number of predictors
```

## Build Neural Network as a function

```python
def simple_network():
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))

    model.compile(optimizer='adam',
            loss='mean_squared_error',
            metrics=['accuracy'])
    return model

model = simple_network()
```

```python
#1 Randomly split the data into a training and test sets by holding 30% of the data for testing. 
#You can use the train_test_split helper function from Scikit-learn.
x_train, x_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3)
```

```python
#2 Train the model on the training data using 50 epochs.
model.fit(x_train, y_train, epochs=50)
```

```python
#3_Evaluate the model on the test data and compute the mean squared error
# between the predicted concrete strength and the actual concrete strength.
# You can use the mean_squared_error function from Scikit-learn.

y_pred = model.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
#print(mse)
#MSE.append(mse)
```

```python
MSE = []
```

## Repeat steps 1 - 3, 50 times, i.e., create a list of 50 mean squared errors.
## Report the mean and the standard deviation of the mean squared errors.

```python
for v in range(51):
    x_train, x_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3)
    model.fit(x_train, y_train, epochs=50, verbose=2)
    y_pred = model.predict(x_test)
    mse = mean_squared_error(y_test, y_pred)
    MSE.append(mse)
```

```python
print('Mean: {}', 'Std_DevL {}, of MSE~ ', np.mean(MSE),',' ,np.std(MSE))
```
